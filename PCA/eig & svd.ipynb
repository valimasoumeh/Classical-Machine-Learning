```python
import matplotlib.pyplot as plt
import numpy as np
from sklearn.datasets import make_classification
```

```python
X,y = make_classification(n_samples=100, n_features=4, random_state=42)
```

```python
plt.scatter(X[:,0], X[:,1], c=y)
```

```python
x_mean = X.mean(axis=0)
x2 = X - x_mean
```

```python
_, mags, vecs = np.linalg.svd(x2)
x_svd = x2 @ vecs[:2,:].T
```

```python
sigma = x2.T @ x2
eigs, eigs_vecs = np.linalg.eig(sigma)
orders = np.argsort(eigs)[::-1]
x_eig = x2 @ eigs_vecs[:,orders]
```

```python
plt.scatter(x_svd[y==1,0], x_svd[y==1,1], c='r', marker='^')
plt.scatter(x_svd[y==0,0], x_svd[y==0,1], c='g', marker='^')
plt.scatter(-x_eig[y==1,0], -x_eig[y==1,1], c='r', marker='+')
plt.scatter(-x_eig[y==0,0], -x_eig[y==0,1], c='g', marker='+')
```
